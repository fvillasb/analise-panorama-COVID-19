# Panorama da COVID-19 no Brasil

A COVID-19 é uma doença infecciosa causada pelo coronavírus SARS-CoV-2. Foi identificada pela primeira vez em Wuhan, na China, em dezembro de 2019 e rapidamente se espalhou pelo mundo, causando uma pandemia global.

Os sintomas mais comuns incluem febre, tosse e falta de ar, mas algumas pessoas podem ter sintomas leves ou nenhum sintoma. A doença pode ser grave e levar a pneumonia, insuficiência dos órgãos e até a morte, especialmente em pessoas idosas ou com condições médicas pré-existentes.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>


A transmissão da COVID-19 ocorre principalmente através do contato com gotículas respiratórias de uma pessoa infectada, como através de tosse ou espirro. Também pode ser transmitida ao tocar em superfícies contaminadas e depois tocando no rosto.

As medidas de distanciamento social, como o uso de máscaras, evitar aglomerações e manter distância física de outras pessoas, são recomendadas para controlar a propagação da doença.

E a vacinação é considerada uma das melhores formas de prevenir a doença e proteger as pessoas vulneráveis.   

É importante destacar que a COVID-19 é uma doença muito contagiosa e sua evolução e tratamento é acompanhado de perto pelas autoridades de saúde e cientistas de todo o mundo.
&nbsp;   
&nbsp;

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados que utilizaremos como fonte foram obtidos por meio do projeto no GitHub que pode ser encontrado [neste link](https://github.com/owid/covid-19-data/tree/master/public/data)

Esse dataset público é uma coleção dos dados sobre a COVID-19 mantidos pela instituição [*Our World in Data*](https://ourworldindata.org/coronavirus)
&nbsp;   
&nbsp;  
A análise de doenças novas como a COVID-19 pode ser desafiadora por vários motivos. Em primeiro lugar, a ciência médica está constantemente evoluindo e os pesquisadores podem levar tempo para entender completamente uma nova doença e como ela se espalha. Em segundo lugar, a COVID-19 é uma doença altamente contagiosa e sua rápida disseminação pode dificultar a coleta de dados precisos e abrangentes.
&nbsp;   

Além disso, pode haver conflitos entre os dados fornecidos pelos governos federal, estaduais e não oficiais. Isso pode acontecer por várias razões, incluindo diferenças na metodologia de coleta de dados, divergências políticas e ideológicas e a falta de transparência em alguns governos.
&nbsp;    

Para lidar com esses desafios, é importante que haja colaboração e comunicação entre autoridades de saúde, pesquisadores e governos em todos os níveis. Também é importante que a coleta de dados seja feita de forma transparente e consistente, para que os pesquisadores e o público possam entender a situação da doença e tomar decisões informadas sobre como se proteger.

## Análise Exploratória dos Dados

A análise exploratória de dados (AED ou EDA - Exploratory Data Analysis) é uma etapa crucial em projetos de Data Science. Ela envolve a exploração e visualização dos dados para extrair insights, entender a distribuição dos dados, identificar padrões e anomalias e detectar possíveis erros nos dados.
&nbsp;   
&nbsp;  
A AED inclui técnicas através das quais é possível identificar relações entre variáveis e detectar quais delas têm maior impacto sobre o resultado esperado.

Trata-se de uma etapa importante porque pode ajudar a identificar problemas nos dados que podem afetar a qualidade dos resultados obtidos a partir da análise. Além disso, ela pode ajudar a orientar as etapas seguintes do projeto, como a seleção de algoritmos de aprendizado de máquina e o ajuste de parâmetros.
&nbsp;  
&nbsp;  
Em resumo, a análise exploratória de dados é uma ferramenta essencial para compreender e preparar os dados para análise mais avançada em projetos de Data Science.


### Análise Exploratória Inicial

Vamos importar os pacotes necessários e iniciar nossa Análise Exploratória.

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# importar o dataset COVID-19 para um DataFrame
DATA_PATH = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df = pd.read_csv(DATA_PATH)
```

#### Características do Dataset

Visualizar o dataset original e obter informações sobre as características dele, tais como dimensões e nomes das variáveis, é fundamental para entendermos com o que estamos lidando.


##### Quais são as dimensões do dataset e as primeiras entradas?


```
# Obter o tamanho do dataset
print("Tamanho do dataset: ",
      '\nVariáveis: ',df.shape[1], "\n"
      'Entradas: ', df.shape[0])

# Visualizar as 5 primeiras entradas do dataset
df.head()
```

##### Quais são os nomes das colunas (variáveis) do dataset?

```
# Extrair e imprimir os nomes das colunas (variáveis) do dataset
for col in df.columns:
  print(col)
```


##### Quais são os tipos de variáveis do dataset?

Obter quais são os tipos atribuidos as variáveis do dataset é essencial para verificarmos se elas estão no formato adequado para as análises a serem feitas. Para isso, utilizaremos o método `dtypes`.

```
# Verificar os tipos das variáveis do dataset
dtypes = dict(df.dtypes) # 'dict' para mostrar todas as variáveis e não se resumir as primeiras e últimas variáveis

for keys, values in dtypes.items():
  print(keys, "=", values)
```

##### Qual é a porcentagem de valores ausentes das colunas?

Obter a quantidade de dados ausentes no dataset é outra informação essencial para a análise exploratória de dados.

Nós precisamos verificar se existem dados faltantes, quais são eles e tentar avaliar os motivos para que isso tenha ocorrido.
Para isso, é importante verificar a fonte dos dados, entender a metodologia de coleta deles, avaliar a aleatoriedade dessa(s) ausência(s), entre outros.

```
# Obter a porcentagem de valores ausentes das colunas
df_ausentes = ((df.isnull().sum() / df.shape[0]) * 100).sort_values(ascending=False)

# Transformar para dicionário para facilitar a impressão dos valores
df_ausentes = dict(df_ausentes)

# Imprimir os resultados
for keys, values in df_ausentes.items():
  print (keys, "= {:.2f}%".format(values))
```

Para melhorar a visualização, vamos plotar um gráfico de barras.

#### Gráfico de barras para visualizar os 5 países com mais mortes

```
# Configurar a visualização dos gráficos
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

# configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'

# definindo cores do gráfico
cores_cinza = CINZA1, CINZA2, CINZA3= "#CFCDCE", "#A2A1A3", "#787878"
cores_azul = AZUL1, AZUL2 = "#8db7eb", "#253760"
```

```
# Obter os 5 países com mais mortes
top5_total_deaths = df.loc[(df.date == data_recente) & (df.continent.notnull()),
                           ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[0:5]

# Criar a figura para o gráfico de barras
fig, ax = plt.subplots(figsize=(8, 4))

# Plotar o gráfico de barras
labels = top5_total_deaths.location
width_bar = 0.24
rects1 = ax.bar('location', 'total_deaths', data=top5_total_deaths, color=cores_azul)

# labels do eixo x
x = np.arange(len(labels))
ax.set_xticks(x)
ax.set_xticklabels(labels)

# # remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# # remover eixo y
plt.gca().axes.get_yaxis().set_visible(False)

# # remover xticks do eixo x
plt.tick_params(
    axis='both',
    which='both',
    bottom=False,
    top=False,
    labelbottom=True,
    colors=CINZA2)

#criar função para dar labels automaticamente
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(
                    '{:,.0f}'.format(height),
                    xy=(rect.get_x() + rect.get_width()/2,  height),
                    xytext=(0, -20),  # offset
                    textcoords="offset points",
                    ha='center',
                    va='bottom',
                    fontsize=10,
                    color="white",
                    )

# plotar labels dentro das barras
autolabel(rects1)

# plotar subtítulo
plt.title("5 regiões ou países com o maior número de mortes",
         fontsize=12,
         loc='left',
         color=CINZA2,
         pad=12)

# plotar título
plt.suptitle(
         "Número de Mortes por Covid-19",
         fontsize=16,
         color=CINZA3,
         ha='right',
         va='baseline',
         fontweight='bold')

# plotar o gráfico
fig.tight_layout();
plt.show();
```

### Como o vírus está progredindo?

```
# Obter as entradas total_deaths associadas à localização World
total_deaths_world = df.loc[df.location == "World", ["total_deaths", "date"]].reset_index(drop=True)

# Criar uma figura para o plot
fig, ax = plt.subplots(figsize=(8, 4))

# Plotar o grafico de linhas
ax.plot('date', 'total_deaths', data=total_deaths_world)

# Personalizar tick e labels do eixo
plt.tick_params(
    axis='both',
    which='both',
    bottom=True,
    top=False,
    labelbottom=True,
    labelsize=9,
    colors=CINZA2)

plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

# Definir o formato da data para melhor visualização
date_form = plt.matplotlib.dates.DateFormatter("%b. %Y")
ax.xaxis.set_major_formatter(date_form)

# # remover spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# incluir grades
ax.grid(linestyle="--", linewidth=0.5, color=CINZA1, alpha=0.5)

# plotar subtítulo
plt.title("Evolução do número de casos no mundo",
         fontsize=12,
         loc='left',
         color=CINZA2,
         pad=12)

# plotar título
plt.suptitle(
         "Número de Casos de Covid-19",
         fontsize=16,
         color=CINZA3,
         ha='right',
         va='baseline',
         fontweight='bold')

# plotar o gráfico
fig.autofmt_xdate()
plt.tight_layout()
```
